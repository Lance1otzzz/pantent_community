```python
import re

# 正则表达式匹配专利号的模式
patent_pattern = re.compile(r'US-\d{7,}-[A-Z]')

# 读取文件内容
def extract_patent_numbers(file_path):
    with open(file_path, 'r',encoding='utf-8') as file:
        text = file.read()

    # 查找所有匹配的专利号
    patent_numbers = patent_pattern.findall(text)

    return patent_numbers

# 假设你的文件路径是 'path_to_your_file.txt'
file_path = 'basf_se.txt'
patents = extract_patent_numbers(file_path)

# 打印所有提取的专利号
for patent in patents:
    print(patent)
```

```python
import re

# 正则表达式匹配专利号的模式
patent_pattern = re.compile(r'US-(\d{7,})-[A-Z]')

# 读取文件内容，并指定编码为 UTF-8
def extract_patent_numbers(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()

    # 查找所有匹配的专利号，并捕获数字部分
    patent_numbers = patent_pattern.findall(text)

    return patent_numbers

# 将提取的专利号数字部分保存到文本文档
def save_numbers_to_file(numbers, output_file_path):
    with open(output_file_path, 'w', encoding='utf-8') as file:
        for number in numbers:
            file.write(number + '\n')

# 假设你的文件路径是 'path_to_your_file.txt'
file_path = 'nelson.txt'
# 假设你想要保存结果到这个文件 'patent_numbers.txt'
output_file_path = 'nelson_extract.txt'

# 提取专利号中的数字部分
patents_numbers = extract_patent_numbers(file_path)
# 保存到文件
save_numbers_to_file(patents_numbers, output_file_path)

print(f"专利号数字部分已保存到 '{output_file_path}'")
```
